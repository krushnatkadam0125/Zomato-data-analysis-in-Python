# Top 10 Restaurants location in Pune

```python
df_locations = df[['dine_rating','loc']]
df_locations
```

```python
best_rest=df_locations[df_locations['dine_rating']>4.2]
best_rest.head(20)
```

```python
best_rest.shape
```

```python
a=best_rest.groupby('loc')
print(a)
```

```python
location_wise_best_restuarants=a.size()
print(location_wise_best_restuarants)
print(type(location_wise_best_restuarants))
```

```python
location_wise_best_restuarants=location_wise_best_restuarants.sort_values(ascending=False)
a=location_wise_best_restuarants[:10]

print(location_wise_best_restuarants)
```

```python
# Bar Plot

plt.figure(figsize=(6,4))
sns.barplot(x=a,y=a.index)
```

```python
# total number of restaurants by location

Number_of_location=df[['loc','dine_rating']]
a=Number_of_location.groupby("loc")
rest_per_location=a.size()
```

```python
rest_per_location=rest_per_location.sort_values(ascending=False)
rest_per_location
```

```python
rest_per_location[:10]
```

```python
# Now we wants to find 
# (Number of rest which got more than 4.2 / total number of rest) per Location
# We are concatinating two series together 

frames=[rest_per_location,location_wise_best_restuarants]
data_n=pd.concat(
        frames,
        axis=1,
        join='inner',
        ignore_index=False,
        copy=True,)
```

```python
data_n
```

```python
data_n[0]
```

```python
print(len(data_n[0].values))
```

```python
data_n['popularity_percentage']=list(map(lambda t: (t[1]/t[0])*100,zip(data_n[0],data_n[1])))
```

```python
data_n
```

```python
data_updated=data_n[data_n[0]>100]
```

```python
data_updated
```

```python
data_updated=data_updated.sort_values(by="popularity_percentage",ascending=False)
```

```python
data_updated
```

```python
sns.barplot(x=data_updated["popularity_percentage"],y=data_updated.index)
```

These are the best location in pune where find the high rated restaurants 

Koregaun Park has the highest popularity percentage out of all the location in pune
